```python
#importing the Libraies
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
dataset
```

```python
#to known the columns names
dataset.columns
```

```python
#convert the data from categorical to numerical data then only we can process the data and first create dummies and then drop the dummies data.
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
indepentent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
indepentent
```

```python
#split the output data and print the data
depentent=dataset[['Profit']]
depentent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indepentent, depentent, test_size = 1/3, random_state = 0)
```

```python
#inthis model creation of decision_tree
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best',max_features='log2')
regressor=regressor.fit(X_train,y_train)
```

```python
#show the tree with malpoltlib
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
